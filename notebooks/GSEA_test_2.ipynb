### Try with a real data

```python
from pals.pimp_tools import get_pimp_API_token_from_env, PIMP_HOST, download_from_pimp
from pals.evaluation import _select_significant_entries, _compute_prec_rec_f1
from pals.common import save_obj, set_log_level_debug, set_log_level_info, set_log_level_warning, DATABASE_PIMP_KEGG, SIGNIFICANT_THRESHOLD
from pals.feature_extraction import DataSource

from loguru import logger
```

```python
token = get_pimp_API_token_from_env()
```

```python
analysis_id_plasma = 636
int_df_plasma, annotation_df_plasma, experimental_design_plasma = download_from_pimp(token, PIMP_HOST, analysis_id_plasma, 'kegg')
```

```python
database_name = DATABASE_PIMP_KEGG
ds_plasma = DataSource(int_df_plasma, annotation_df_plasma, experimental_design_plasma, database_name)
```

```python
standardize = True
```

```python
logger.debug('Calculating GSEA')
measurement_df = ds_plasma.change_zero_peak_ints()
if standardize:
    scaled_data = np.log(np.array(measurement_df))

    # Put the scaled data back into df for further use
    sample_names = measurement_df.columns
    measurement_df[sample_names] = scaled_data

D = measurement_df.values
```

```python
D
```

```python
annot_df = ds_plasma.get_annotations()
annot_df
```

```python
measurement_df
```

```python
joined = pd.merge(how='inner', left=measurement_df, right=annot_df, left_index=True, right_index=True)
joined = joined.set_index('entity_id')
unique_ids = [ds_plasma._get_unique_id(x) for x in joined.index.values]
joined.index = unique_ids
joined
```

```python
# gene_sets is a list of lists.
# Each list in gene_sets contains indices of rows in measurement_df for that pathway.
gene_sets = {}
assert len(ds_plasma.dataset_pathways) > 0, 'No pathways found in the dataset'
pathways = list(ds_plasma.dataset_pathways)
for pw in pathways:
    pathway_row_ids = ds_plasma.dataset_pathways_to_row_ids[pw]
    pw_unique_ids = []
    for row_id in pathway_row_ids:
        pw_unique_ids.extend(ds_plasma.dataset_row_id_to_unique_ids[row_id])
    pw_unique_ids = list(set(pw_unique_ids))
    gene_sets[pw] = pw_unique_ids
```

```python
comp = ds_plasma.comparisons[0]
comp
```

```python
case = comp['case']
control = comp['control']
logger.debug('Running comparison case=%s control=%s' % (case, control))
pheno_cols = set(ds_plasma.get_experimental_design()['groups'][case])
df_cols = measurement_df.columns.values
```

```python
C = []
for col in df_cols:
    if col in pheno_cols:
        C.append(1)
    else:
        C.append(0)
C = np.array(C)
```

```python
data = joined
gene_sets = gene_sets
cls = C.tolist()
outdir = None
min_size = 2
max_size = 500
permutation_num = 1000
weighted_score_type=1
permutation_type='phenotype'
method='log2_ratio_of_classes'
ascending=True
processes=1
figsize=(6.5,6)
format='pdf',
graph_num=20
no_plot=True
seed=None
verbose=True
```

```python
gs = GSEA(data, gene_sets, cls, outdir, min_size, max_size, permutation_num,
          weighted_score_type, permutation_type, method, ascending, processes,
           figsize, format, graph_num, no_plot, seed, verbose)
gs.run()
```

```python
df = gs.res2d
pw_names = [ds_plasma.pathway_dict[pw]['display_name'] for pw in df.index.values]
df.insert(0, 'pw_name', pw_names)
df = df.reset_index()
```

```python
selected = df[['Term', 'pw_name', 'pval', 'fdr', 'es']]
selected = selected.rename(columns={'Term': 'mapids'}).set_index('mapids')
```
